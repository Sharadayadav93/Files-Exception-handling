```python
Theory question

1.What is the difference between interpreted and compiled languages?
       Compiled Languages:
       Compilation Process: In compiled languages, the source code is transformed
       into machine code (binary code) by a compiler before it is run on a computer.
       This machine code is what the processor understands.
      Execution: Once compiled, the program is executed directly by the computer,
      meaning it runs faster, as the translation to machine code has already occurred.
      Examples: C, C++, Rust, Go, Fortran.

      Interpreted Languages:
      Interpretation Process: In interpreted languages, the interpreter reads
       source code line-by-line and executes it directly, without transforming
       it into machine code first. The translation happens during execution.
       Execution: The program runs slower since it has to be interpreted every
       time it is executed. The source code is not converted to machine code
       beforehand.
       Examples: Python, JavaScript, Ruby, PHP

 2.  What is exception handling in Python?
        Exception handling in Python is a mechanism that allows you to deal with
         errors (exceptions) that occur during the execution of a program,
         in a controlled way. When an error occurs, Python generates an exception
          (an object that describes the error) and can either stop the program
           or allow you to handle the error gracefully using specific code.

      Key Concepts in Python Exception Handling:
      Try Block: The code that might raise an exception is placed inside a try block.
      Except Block: If an exception occurs inside the try block, the program
      jumps to the except block to handle the exception.
      Else Block: If no exception occurs in the try block, the code inside the
      else block will run. This is optional.
      Finally Block: Code inside the finally block is always executed,
      whether an exception occurs or not. It's typically used for cleanup
      actions, like closing files or releasing resources.

 3.What is the purpose of the finally block in exception handling?

         The finally block in exception handling is used to define code that
         should be executed no matter what—whether an exception was raised
         or not. This block is typically used for cleanup operations that need
          to happen regardless of whether the program encountered an error or
          ran successfully.

4.What is logging in Python?

     It Records the state and flows of your program/software. it is useful for
     understanding, monitoring and debugging your code.it show how program behav
     es overtime.

5.What is the significance of the __del__ method in Python?

        the __del__ method is useful for cleanup tasks when an object is
         destroyed or garbage collected. However, due to the unpredictable
         timing of garbage collection, it's often recommended to use other
         mechanisms like context managers or explicit cleanup functions
         for managing resources.

6.What is the difference between import and from ... import in Python?

       import:
       The entire module is imported.
       Access components with module prefix (e.g., module_name.function())
       The module itself is in the namespace.
       Ex - import math
             math.sqrt(16)

       From Import:
       Specific functions, classes, or variables from the module is imported.
       Access components directly (e.g., function())
       The specific components are directly in the namespace
       Ex. from math import sqrt
            sqrt(16)

7. How can you handle multiple exceptions in Python?

       Use multiple except blocks for handling different exceptions individually.
       Use a tuple in a single except block to handle multiple exceptions at once.
        Catch all exceptions using Exception, but it’s generally better to be
        more specific about what you handle.Combine else and finally to refine
        your control flow.Re-raise exceptions if you need to handle them partially
         and then propagate them.

8. What is the purpose of the with statement when handling files in Python?

       The with statement in Python is used to simplify resource management,
       especially when working with files. It is part of Python's
       context management system. The purpose of using the with statement
       when handling files is:No need to explicitly close the file,
       it is automatically closed when the block end.

9.What is the difference between multithreading and multiprocessing?

       Concurrency Model:
       Multithreading: Involves running multiple threads within the same process.
       Threads share the same memory space and can access shared data directly.
       This is suitable for I/O-bound tasks (e.g., reading/writing to a file,
       making network requests), where tasks spend a lot of time waiting for
       external resources.
       Multiprocessing: Involves running multiple processes, each with
       its own memory space and resources. Processes are independent of
       each other and do not share memory directly, making them suitable
       for CPU-bound tasks (e.g., complex computations) where you want to
       fully utilize multiple CPU cores.


10. What are the advantages of using logging in a program?

      Structured output with different log levels.
      Better control over where and how logs are output
       (e.g., console, files, remote servers).
      Easier debugging by providing persistent logs with timestamps.
      Separation of concerns, keeping your code clean and maintainable.
      Flexibility to change logging configurations without modifying code.
      Thread-safe, supporting concurrent programs.
      Useful for monitoring, auditing, and alerting in production.


11. What is memory management in Python?

       Memory management in Python is largely automatic. Python handles
        memory allocation and deallocation using reference counting and
        garbage collection, making it easier for developers to focus on
        writing code without worrying too much about memory management.
        However, understanding these mechanisms can help you write more
        efficient programs and debug memory-related issues when they arise.

12.  What are the basic steps involved in exception handling in Python?

        Try Block: Attempt to execute code that might raise an exception.
        Except Block: Handle specific exceptions or errors if they occur.
        Else Block: Optionally, execute code if no exception occurs.
        Finally Block: Optionally, execute code that should run no matter what,
         typically for clean-up


13.Why is memory management important in Python?

        Memory management in Python is crucial for several reasons,
        and understanding it helps ensure that your programs run efficiently,
        are reliable, and do not run into issues like memory leaks or excessive
        resource consumption. Here’s why memory management is important in Python

14.  What is the role of try and except in exception handling?

          Try block allows you to identify and isolate the part of your
           program where an exception might occur, so you can address
           potential issues proactively.

           The except block allows you to define custom handling for errors,
           such as logging, showing a user-friendly message, or attempting to
           recover from the error.

 15. How does Python's garbage collection system work?

          Reference Counting:
          Python automatically tracks the number of references to objects.
          When an object's reference count drops to zero, the memory is freed.

          Cyclic Garbage Collection:
          Python's garbage collector identifies and cleans up cyclic references
           (where objects reference each other) that reference counting cannot handle.

          Generational Garbage Collection:
          Objects are grouped into generations based on their age.
          Younger objects are collected more frequently, while older
          objects are collected less frequently.

           Manual Control:
           You can manually control garbage collection using the gc module,
            enabling, disabling, or forcing collection.

           Weak References:
           Python provides weak references to avoid preventing objects from
           being collected, useful in certain design patterns.

16.  What is the purpose of the else block in exception handling?

          The else block in exception handling in Python is used to define code
          that should only run if no exception was raised in the try block.
          It provides a way to execute code after the try block succeeds
          without encountering any errors.

17. What are the common logging levels in Python?

         Info - General information about the program's operation.
         Warning - Indication of a potential problem or unexpected behavior.
         DEbug - Detailed information for debugging, very verbose.
         Error -Errors that might impact functionality but don't stop the program.
         Critical - Critical errors that typically stop program execution.

18. What is the difference between os.fork() and multiprocessing in Python?

      os.fork() is a low-level system call for creating processes in Unix-like
      environments. It is fast and lightweight but requires manual
      handling of process synchronization and communication, and is not cross-platform.


      multiprocessing is a high-level, cross-platform module that
      simplifies the creation, management, and synchronization of processes.
       It is ideal for parallel and concurrent programming and provides
       a more flexible and robust approach for working with multiple processes.

 19.What is the importance of closing a file in Python?

       In Python, closing a file is important to:
       >Release system resources (file handles).
       >Ensure all data is written and saved properly.
       >Avoid data corruption or loss.
       >Prevent file access conflicts or locking issues.
      >Achieve cleaner, more maintainable code, especially when using the with state

 20.What is the difference between file.read() and file.readline() in Python?

       file.read():
       Purpose: Reads the entire content of the file at once.
       Use Case: Use file.read() when you need to process the entire content
        of the file at once.


        file.readline():
        Purpose: Reads the next line from the file.
        Use Case: Use file.readline() when you need to read the file line by line,
        especially for large files or when you only need to process one line at a time.

 21. What is the logging module in Python used for?

          The logging module in Python is used to log messages from your
          application, which helps track events, errors, warnings, and
          other information that might be helpful for debugging, monitoring,
           or understanding the flow of a program. Logging provides a systematic
           way to record the state of the program and its behavior during execution.

 22.What is the os module in Python used for in file handling?


The os module in Python is a standard library that provides a way to interact
 with the operating system. It includes functionalities for file handling,
 such as interacting with file paths, directories, and performing file-related
 operations.In the context of file handling, the os module provides several
 methods that allow you to manage and manipulate files and directories in a
 platform-independent way.

23. What are the challenges associated with memory management in Python?

      Use gc.collect(): Manually trigger garbage collection when needed
       (e.g., in long-running applications) to reclaim memory.
      Avoid Circular References: Minimize the use of circular references to
      reduce the chances of memory leaks.
      Use Memory-Efficient Data Structures: Use libraries like NumPy for
      large arrays, or collections.deque for efficient queue operations.
      Be Cautious with Global Variables: Global variables can unintentionally
      retain memory, so use them sparingly.
      Profile Memory Usage: Use memory profiling tools (e.g., memory_profiler,
      objgraph, guppy3) to identify memory usage patterns and potential leaks.
```
